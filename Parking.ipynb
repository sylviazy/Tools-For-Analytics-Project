```python
# read dataframe
import pandas as pd
df = pd.read_csv('311_Service_Requests_2020.csv')
df
```

```python
# see what's the data in 'incident Zip'
df['Incident Zip'].unique()
```

```python
# get df in my zip: 10036
df_10036 = df[ df['Incident Zip'] == 10036]
df_10036
```

```python
# see what types of incidents we have
df_10036['Complaint Type'].unique()
```

```python
# fraction of parking incidents in my area
parking_fraction_10036 = sum(df_10036['Complaint Type'] == 'Illegal Parking') / df_10036.shape[0]
parking_fraction_10036
```

```python
# fraction of parking incidents global
parking_fraction_glob = sum(df['Complaint Type'] == 'Illegal Parking') / df.shape[0]
parking_fraction_glob
```

```python
# compute 'higher_parking_proportion'
higher_parking_proportion = parking_friction_10036 > parking_friction_glob
higher_parking_proportion
```
